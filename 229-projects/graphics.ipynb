* * * * *

A plot is nothing more than a connect-the-dot graph of $x$ and $y$
values. It can be useful to know how to do the steps. The above could be
done with:



```julia
a, b = 0, 2pi
xs = linspace(a, b)     # 100 points between a and b
ys = map(sin, xs)           # or ys = [sin(x) for x in xs] (see the notes)
plot(x=xs, y=ys, Geom.line) # Geom.line connects the points. Try without to see.
```

The `xs` and `ys` are written as though they are "plural" because these
variables contain 100 values each in a container (a vector in this
case). The `map` command "maps" a function to each value in the
container and returns a new container with the "mapped" values. In the
example above, these are the values for the `sin` at each `x`.
Containers (vectors in this case) are often constructed by combining
like values within square brackets separated by commas: e.g., `[a,b]`.
For plotting, we can combine functions using `[]` and all will plot:



```julia
plot([sin, cos], 0, 2pi)
```
