This project covers some of the techniques of integration utilizing the
symbolic algebra capabilities provide by the `SymPy` package.
The basic techniques of integration involve

-   substitution (reverse chain rule)
-   integration by parts (reversed product rule)
-   partial fractions (for rational functions)

The `SymPy` function `integrate` is a implementation of the [Risch
Algorithm](http://en.wikipedia.org/wiki/Risch_algorithm). This algorithm
can decide if a function comprised of elementary functions has an
antiderivative expressed in elementary function. For example, neither
$\sin(x^2)$ or $\e^{x^2}$ has such a "nice" antiderivative, whereas
$\sin(x)^2$ and $xe^{-x^2}$ do. Risch's algorithme exploits [Liouville's
Theorem](http://asmeurersympy.wordpress.com/2010/08/) which gives an
representation for any potential antiderivative of $f(x)$ using
*elementary functions* such as polynomials, logarithms, exponentials and
trigonometric functions.

### It doesn't always work ...
The `integrate` function works for many functions, but for some one can
help by performing substitutions by hand first.

Here is an example from
[here](http://faculty.uml.edu/jpropp/142/Integration.pdf) where
performing the substitution by hand is needed to give a boost:



```julia
x, dx = symbols("x, dx")
f(x) = sqrt(x^2 + 1)
u = x*log(x)
ex = f(u) * diff(u,x)
integrate(ex, x)
```

Sympy can't do this integral, and so returns the expression. But this is
really ripe for substitution with $u = x\log(x)$.
Let's introduce a variable `dx` into our expression, and then we have:



```julia
ex =  f(u) * diff(u,x) * dx
```

Then we can have this substitution $u = x*log(x)$ (so
$du = (x\cdot\log(x))' dx$) implemented in `julia` via:


```julia
u, du = symbols("u, du")
## replace x*log(x) with u and diff(x*log(x),x) * dx with du
ex1 = subs(ex, x*log(x), u) |> subs(diff(x*log(x), x) * dx, du)
```

This latter function can be integrated (after stripping off the
differential we added for familiarity):


```julia
ex = integrate(ex1 / du)
```

If needed we can resubstitute in $x \log(x)$ for $u$ to get an answer.


```julia
subs(ex, u, x*log(x))
```

### Integration of rational functions:
Some background reading on the implementation is
[here](http://asmeurersympy.wordpress.com/2010/06/11/integration-of-rational-functions/)

A *rational function* is a ratio of polynomial functions. Using
polynomial long division and dividing out common factors, up to
removable singularities, one can uniquely write a rational function in
terms of three other polynomials:

$$
a(x)/b(x) =  s(x) + r(x)/q(x)
$$

where the degree of $r(x)$ is less than the degree of $q(x)$ and there
are no common roots to $r(x)$ and $q(x)$. It is easy to integrate
$s(x)$. What about the $r(x)/q(x)$?

We know from the *fundamental theorem of algebra* that we can factor
$q(x) = p_1(x)^n_1 \cdots p_k(x)^{n_k}$, where $p_i$ is a linear or
quadratic factor. Further, the [partial fraction
decomposition](http://en.wikipedia.org/wiki/Partial_fraction_decomposition)
ensures then that the ratio can be written as:

$$
\frac{r(x)}{q(x)} = \sum_i^k (\frac{a_{i1}(x)}{p_i(x)} + \frac{a_{i2}(x)}{p_i(x)^2} + cdots + \frac{a_{in_i}(x)}{p_i(x)^n_i}
$$

with $a_{ij}(x)$ being polynomials with degree less than the $p_i(x)$,
so in this case either a constant of linear polynomial.

So, if the polynomials of the form $(ax+b)/(cx^2 + dx + e)^j$ can be
integrated, by the linearity of integration all rational functions can
be integrated. Let's investigate this question.

```julia
function usub(ex, fx)
    u, du, dx = symbols("u, du, dx")
    ex1 = ex * dx
    ex2 = subs(ex1, dx, du/diff(fx, x))
    ex3 = subs(ex2, fx, u)
    ex3 / du
end

ex = log(log(x))/x
usub(ex, log(x))
        
```

```julia
function uvparts(ex, let_u_equal)
    u, dv = let_u_equal, ex/let_u_equal
    du = diff(u, x)
    v = integrate(dv)
    [u*v, v*du]
end

ex = x*sin(x)
uv, vdu = uvparts(ex, sin(x))
```

```julia
function polar(fs::Tuple, a, b)
    ts = linspace(a, b)
    f,g = fs
    xs, ys = map(f, ts), map(g, ts)
    plot(x=xs, y=ys, Geom.line(preserve_order=true))
end

polar((sin, x->cos(3x)), 0, 2pi)
```

```julia
function polar(r, a, b)
    x(theta) = r(theta) * cos(theta)
    y(theta) = r(theta) * sin(theta)
    thetas = linspace(a, b, 250)
    xs, ys = map(x, thetas), map(y, thetas)
    plot(x=xs, y=ys, Geom.line(preserve_order=true))
end

polar(t -> cos(2t), 0, 2pi)
```
